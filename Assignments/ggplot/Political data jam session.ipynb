```python
import pandas as pd
from ggplot import *
%matplotlib inline

df = pd.read_csv('../../Data/commercials.csv')

df.head()
```

```python
candidates = df[(df["candidates"] =="Hillary Clinton") | (df["candidates"] =="Donald Trump")]
```

```python
procon = candidates[(candidates["message"] == "pro") | (candidates["message"] == "con")]

procon.head()
```

```python
ggplot(aes(x='message', color='candidates'), data=candidates) + \
     geom_bar() + facet_grid('candidates') + facet_grid('candidates')
```

```python
clinton = df[(df["candidates"] =="Hillary Clinton")]
```

```python
trump = df[(df["candidates"] =="Donald Trump")]
```

```python
ggplot(aes(x='message', fill='message'), data=clinton) + \
     geom_bar()
```

```python
ggplot(aes(x='message', fill='message'), data=trump) + \
     geom_bar(binwidth=100)
```

```python
ggplot(aes(x='message'), data=procon) + \
     geom_bar() + facet_wrap(x='location', y='candidates')
```
