And now, taking the ggplot code we wrote in Python and pasting it in here I get ...

And error message. But the reason is simple: In R, you don't have to put your field names in quotes. THAT'S LITERALLY IT. THAT'S THE TOTAL DIFFERENCE.

```r
ggplot(enrollment, aes(x=Year, y=Enrollment)) + geom_line(color="red", size = 2) + theme_bw()
```

Same labelling. 

```r
ggplot(enrollment, aes(x=Year, y=Enrollment)) + geom_line() + labs(x="Academic year", y="Fall enrollment", title="Enrollment at UNL since 1967")
```

Scales are *slightly* different in that in python, you create a tuple of values like this: (10000, 30000). In R, you define a *vector* by putting a c in front of your (). So like this:

```r
ggplot(enrollment, aes(x=Year, y=Enrollment)) + geom_line() + labs(x="Academic year", y="Fall enrollment", title="Enrollment at UNL since 1967") + scale_y_continuous(limits = c(1000, 30000)) + scale_x_continuous(limits = c(1967, 2016)) 
```

The same themes work. 

```r
ggplot(enrollment, aes(x=Year, y=Enrollment)) + geom_line() + labs(x="Academic year", y="Fall enrollment", title="Enrollment at UNL since 1967") + scale_y_continuous(limits = c(1000, 30000)) + scale_x_continuous(limits = c(1967, 2016)) + theme_bw() 
```

And as an added bonus, you can save your work out as a PDF and open it in Illustrator. Each element of the graph is a vector in Illustrator. So whatever you can't fix in ggplot, you can in Illustrator. 

```r
ggsave("enrollment.pdf", device = "pdf")
```
