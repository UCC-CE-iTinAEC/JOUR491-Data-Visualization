```r
library(dplyr)
library(ggplot2)
```

```r
voters <- read.csv("../../Data/map.csv")
```

```r
head(voters)
```

```r
ggplot(aes(x=Map), data=voters) + geom_bar() + coord_flip() + labs(x = "", y = "Number of Counties", title = "Nebraska's shrinking counties becoming more Republican") + geom_text(aes(label = ..count.., y= ..prop..), stat= "count", hjust = -.5)
```
